# Histogramas de frecuencias

```python
clases = int(np.ceil(1+np.log2(3333)))  # Regla de Sturges
plt.hist(data2['Day Calls'], bins=clases) # bins=[0,25,50,75,100,125 ....200]
plt.xlabel('Nro de llamadas al día')
plt.ylabel('Frecuencia')
plt.title('Histograma del Nº de llamadas al día')
```

```python
plt.boxplot(data2['Day Calls'])
plt.ylabel('Nro de llamadas al Día')
plt.title('Boxplot de llamadas al Día')
```
